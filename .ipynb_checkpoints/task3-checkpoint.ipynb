```python
from numpy import asarray
plt.figure(figsize=(15,15))
# Read the original image
img=resized_images[105]
img=asarray(img)

# Blur the image for better edge detection
img_blur = cv2.GaussianBlur(img, (3,3), 0) 
 
# Sobel Edge Detection
sobelx = cv2.Sobel(src=img_blur, ddepth=cv2.CV_16S, dx=1, dy=0, ksize=3) # Sobel Edge Detection on the X axis
sobely = cv2.Sobel(src=img_blur, ddepth=cv2.CV_16S, dx=0, dy=1, ksize=3) # Sobel Edge Detection on the Y axis
sobelxy = cv2.Sobel(src=img_blur, ddepth=cv2.CV_16S, dx=1, dy=1, ksize=5) # Combined X and Y Sobel Edge Detection
# Laplacian Edge detection
laplacian = cv2.Laplacian(img,ksize=3, ddepth=cv2.CV_16S)
# Canny Edge Detection
canny = cv2.Canny(image=img_blur, threshold1=100, threshold2=200) # Canny Edge Detection

plt.subplot(3,2,1),plt.imshow(img)
plt.title('Original'), plt.xticks([]), plt.yticks([])

plt.subplot(3,2,2),plt.imshow(laplacian)
plt.title('Laplacian'), plt.xticks([]), plt.yticks([])

plt.subplot(3,2,3),plt.imshow(canny)
plt.title('Canny'), plt.xticks([]), plt.yticks([])

plt.subplot(3,2,4),plt.imshow(sobelxy)
plt.title('Sobel XY'), plt.xticks([]), plt.yticks([])

plt.subplot(3,2,5),plt.imshow(sobelx)
plt.title('Sobel X'), plt.xticks([]), plt.yticks([])

plt.subplot(3,2,6),plt.imshow(sobely)
plt.title('Sobel Y'), plt.xticks([]), plt.yticks([]);
```

```python
def load_images_from_dir(dir_location, filenames):
    return [np.array(Image.open('{}/{}'.format(dir_location, filename)).resize((64, 64))) for filename in filenames]
    
images = np.array(load_images_from_dir(train, filenames))
images.shape

# Dimesion Reduction
# Pixel * Pixel * RGB
print(images[0].shape)
h, w, d = images[0].shape
images_resized = np.array([np.reshape(img, (w*h*d)) for img in images])
images_resized.shape
```

```python
print(images[0].shape)
h, w, d = images[0].shape
images_resized = np.array([np.reshape(img, (w*h*d)) for img in images])
print(images_resized.shape)
```

```python
from sklearn.tree import plot_tree
from sklearn.preprocessing import PolynomialFeatures
from sklearn.tree import DecisionTreeClassifier

tree = DecisionTreeClassifier(max_depth=4)
tree.fit(images_resized, labels)
```

```python
tree.score(images_resized, labels)
```

```python
plot_tree(tree)
```

```python
from sklearn.ensemble import RandomForestClassifier
forest = RandomForestClassifier()
forest.fit(images_resized, labels)
```

```python
forest.score(images_resized, labels)
```

```python
test_filenames = os.listdir(test1)
test_images = np.array(load_images_from_dir(test1, test_filenames))
test_images.shape
```

```python
test_images_resized = np.array([np.reshape(img, (w*h*d)) for img in test_images])
test_images_resized.shape
```
